# productage가 1보다 작으면1 
# 2보다 작으면 2 
# 3보다 작으면 3 
# 이외는 5라고 새로운 컬럼 product_age_new 를생성하세요

```python
csData = pd.read_csv("../dataset/customerdata.csv")
csData # 데이터 불러오기
```

```python
def age(value):
    if value < 1:
        return 1
    elif value < 2 :
        return 2
    elif value < 3:
        return 3
    else:
        return 5
#함수를 만들어 조건을 정의하기   
```

```python
csData["product_age_new"] = csData.PRODUCTAGE.apply(age)
#만들어 둔 함수를 이용하여 PRODUCTAGE컬럼을 조건으로 조회
```

# [데이터통합]
# selloutData 자료에서 YEAR, WEEK 컬럼을 생성하고 
# WEEK가 52이하인 데이터만 조회한 후 refinedSelloutData 변수에담으세요

```python
# 1. 인덱스로 조회하는 경우
selloutData = selloutData.astype({'YEARWEEK':str})
selloutData.dtypes # YEARWEEK 컬럼을 문자형으로 바꿔줌
```

```python
selloutData["YEAR"] = selloutData.YEARWEEK.str[:4] # YEARWEEK의 4번째 문자열까지를 YEAR 컬럼으로 만듬
selloutData["WEEK"] = selloutData.YEARWEEK.str[4:] # YEARWEEK의 4번째부터의 문자열까지를 WEEK 컬럼으로 만듬
```

```python
selloutData
```

```python
refinedSelloutData = selloutData.loc[selloutData.WEEK.astype(int) <= 52] # 52이하를 구하기 위해 정수로 바꿔줌
refinedSelloutData
```

```python
# 2. 숫자 형태로 조회하는 경우
selloutData = selloutData.astype({'YEARWEEK':int})
selloutData.dtypes # YEARWEEK 컬럼을 정수형으로 바꿔줌
```

```python
selloutData["YEAR"] = selloutData.YEARWEEK // 100 # YEARWEEK를 100으로 나눈 몫만큼을 YEAR 컬럼으로 재정의
selloutData["WEEK"] = selloutData.YEARWEEK % 100 # YEARWEEK를 100으로 나눈 나머지만큼을 YEAR 컬럼으로 재정의
# 이 경우에는 WEEK 컬럼이 숫자형태로, 0으로 시작할 수 없기 때문에 0이 사라져 버림
```
