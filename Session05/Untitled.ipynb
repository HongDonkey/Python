```python
selloutData["YEAR"] = selloutData.YEARWEEK.str[:4]
selloutData["WEEK"] = selloutData.YEARWEEK.str[4:] 
```

```python
condiData = selloutData.loc[(selloutData.REGIONID == 'A01') &
               (selloutData.YEAR == '2015')]
```

```python
groupData = condiData.groupby(by='PRODUCT')[["QTY"]].agg(["mean"])
```
