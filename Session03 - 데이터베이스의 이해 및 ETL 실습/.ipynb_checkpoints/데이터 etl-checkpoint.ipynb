# 내 postgres DB에 테이블 정보를 저장하고 해당 테이블 정보를 동료의 DB로 보내라

```python
# 내 postgres DB에 저장

import psycopg2            # psycopg2 임포트
from sqlalchemy import create_engine        # create_engine 임포트  

mylocalpgUrl = "postgresql://postgres:postgres@127.0.0.1:5432/postgres"  # 내 데이터 베이스 주소를 변수에 할당
engine_pg = create_engine(mylocalpgUrl)     #  엔진을 내 데이터 베이스 주소로 설정
resultTableName = "sfsaf"    # 데이터베이스에 저장할 테이블 이름을 변수에 할당
data.to_sql(name = resultTableName, con = mylocalpgUrl, if_exists = "replace", index=False )   # 해당 테이블을 sql로 보내라(이름, 주소, 해당테이블이 있다면 "교체", 컬럼X)
sqlReadData = pd.read_sql(sql="select * from pg_out", con=engine_pg)    # pg_out테이블에서 engine_pg를 통해 모든 정보를 불러와라
```

```python
# 재현씨 postgres DB로 보내기

JHUrl = "postgresql://postgres:postgres@192.168.110.10:5432/postgres"      # 재현씨의 DB주소
data.to_sql(name = resultTableName, con = JHUrl, if_exists = "replace", index=False ) #  재현씨 DB로 내 DB에서 테이블을 보냄
```
