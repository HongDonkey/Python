```python
# pip install selenium # 자동화를 위한 셀레니움 설치

import pandas as pd  # pandas import
from selenium import webdriver # selenium import
from bs4 import BeautifulSoup # bs4 import
from selenium.webdriver.common.keys import Keys # 검색하게 해주는 라이브러리
import unicodedata # li 태그 파싱하면서 utf8로 인코드 변환하기 위해 사용

options = webdriver.ChromeOptions()

# 셀레니움 헤더변경 필요
user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:71.0) Gecko/20100101 Firefox/71.0'    
options.add_argument('user-agent={0}'.format(user_agent))


# options.add_argument('headless')
# 드라이버옵션설정
# 헤드리스 옵션 / 일반 옵션
# options.add_argument('--headless')  # 헤드리스 옵션(리눅스 - 화면을 띄우지 않음)
options.add_argument("window-size=1920x1080") # 일반옵션(윈도우서버 - 개발할 때 디버깅 확인용도)
options.add_argument("disable-gpu")

driverLoc = "../PhoneSpec/chromedriver-win64/chromedriver.exe"  # 드라이버위치설정

# 웹드라이버정의
driver = webdriver.Chrome(executable_path=driverLoc,options=options)
# driver = webdriver.Chrome(dirver_loc, options=options) 
#executable_path에 변수를 지정하여 사용햇다면 options에도 변수를 지정하여 사용해야 함
#자동화된 테스트 소프트 웨어에 의헤 제어되는 크롬창 열림

# 웹페이지 파싱될 때까지 최대 3초 기다림
driver.implicitly_wait(3)

# URL 이동
driver.get("https://www.apple.com/kr/iphone/compare") # 애플 아이폰
```

```python
XPath = '//*[@id="main"]/section[1]/p/button'# 모든 iPhone 모델
```

```python
iPhoneAllModel = driver.find_element_by_xpath(XPath) # 모든 iPhone 모델
iPhoneAllModel.click() # 클릭
```

```python
pgSource = driver.page_source  # 페이지의 개발자 정보를 변수에 담음
```

```python
pgObj = BeautifulSoup(pgSource, "html.parser") # 개발자 페이지 코드를 bs4를 이용하여 출력
```

```python
allModels=pgObj.find_all(name = "div", attrs={"class":"device-list"}) 
allModels
```

```python
allItems=pgObj.find_all(name = "div", attrs={"class":"device-item"}) 
allItems
```

```python
itemsName = pgObj.find_all(name = "div", attrs={"class":"typography-label"})
nameList = [] # 아이폰 모델 이름들
for i in range (3, len(itemsName)): # 메인에 3개의 중복된 모델은 제외
    replacedName = unicodedata.normalize('NFKD', itemsName[i].get_text())
    nameList.append(replacedName)
print(nameList)
```

```python
productList = [] # 아이폰 모델별 스펙(태그 제거 전)
for i in range(0, len(allItems)):
    iPhoneUlTag = allItems[i].find(name="ul", attrs={"class":"with-list-bullet"})
#     print(iPhoneUlTag)
#     iPhoneLiTag = allItems[i].find_all(name="li")
#     print(iPhoneLiTag)
    iPhone = unicodedata.normalize('NFKD', iPhoneUlTag.get_text())
    print(iPhone)
    print("===================")
    productList.append(iPhone)
# print(productList)
```

```python
iPhoneInfo = {}
specInfo = []
# print(len(productList[29]))
for i in range(0, len(productList)):
#     iPhoneModel = productList[i].find_all(name="li")
    iPhoneInfo[nameList[i]] = productList[i]
    
# print(nameList)
print(iPhoneInfo)    
# print(iPhoneInfo)

```
