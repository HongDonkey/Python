```python
targetParent = htmlObj.find(name="div", attrs = {"id":"id_spark_quiz"})
targetParent
```

```python
value = [] # 타이틀 값을 넣어줄 리스트
link = [] # 링크 값을 넣어줄 리스트
aTags = targetParent.findAll(name="a") # 태그 A로 시작하는 것을 모두 찾아줌


for i in range(0, len(aTags)):
    title = aTags[i].text     # A태그 안의 내용을 Title변수에 담음
    value.append(title)       # title변수를 value리스트에 추가함
    link2 = aTags[i].attrs["href"]  # A태그 안의 링크를 link2변숭 ㅔ담음
    link.append(link2)              # link2 변수를 link변수에 담음
    
pd.DataFrame(zip (value, link), columns=["퀴즈타이틀","링크"])  # zip을 사용하여 2가지 배열을 하나의 데이터프레임 형식으로 출력함
```

# 동원홈페이지에서 냉동식품 부분의 링크를 가져와 데이터프레임 형식으로 만들기

```python
from bs4 import BeautifulSoup # 태그를 이쁘게 쓰게해준다
import requests # 웹페이지에 접속요청
import requests, bs4
import pandas as pd
```

```python
url = "https://www.dongwonmall.com/category/main.do?cate_id=0111005600010001"     # 데이터를 조회할 url
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:71.0) Gecko/20100101 Firefox/71.0'}
# 나는python으로 접속한게 아니라 user-agent를 바꿔서... Mac에서 접속한.... 일반 사용자다!
data=requests.get(url,headers=headers) 
data.text
resp = requests.get(url)
resp.encoding = "utf-8"
```

```python
html = resp.text
htmlObj = BeautifulSoup(html, "html.parser")


productList = []
priceList = []
product = htmlObj.find_all(name="span",attrs = {"class":"lktxt"})


for i in range(0, len(product)):
    eachProduct=product[i].text
    renameProduct = eachProduct.replace("\n","")
    productList.append(renameProduct)
    productList[i].replace("\n","")


priceInfo = htmlObj.find_all(name="div" ,attrs ={"class":"listtype_prc_wrap"})
for i in range(0,len(productList)):
    prc_new=priceInfo[i].find(name="span", attrs ={"class":"prc_new"})
    emPrice = prc_new.find(name="em")
    price = emPrice.text
    priceList.append(price)
pd.DataFrame(zip (productList, priceList), columns=["상품명","가격"])
```
