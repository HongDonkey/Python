```python
import requests        # API 서비스에 연결
from lxml import html  # parse XML 사용
from bs4 import BeautifulSoup   # 정렬해주는 BeautifulSoup 사용
import pandas as pd
from urllib.request import Request, urlopen
from urllib.parse import urlencode, quote_plus,unquote

API_Key = 'AT33s775KYpJOkUBJu0dxkJuUeIfDIOJRzAH084EQS3JN%2BzFjErLHuk%2FGZa9L4gBTSGCzeA69tI9PwLp7B37IQ%3D%3D' # 서비스 키
url="http://openapi.molit.go.kr/OpenAPI_ToolInstallPackage/service/rest/RTMSOBJSvc/getRTMSDataSvcAptTradeDev" # 실질적인 url 주소
queryParams = '?' + urlencode({          # 아래 3가지 요소를 합한 값
    quote_plus('ServiceKey') : API_Key,  # ServiceKey
    quote_plus('LAWD_CD'): '11110',      # LAWD_CD
    quote_plus('DEAL_YMD'):'201512'      # DEAL_YMD
})

response = requests.get("http://openapi.molit.go.kr/OpenAPI_ToolInstallPackage/service/rest/RTMSOBJSvc/getRTMSDataSvcAptTradeDev?serviceKey=AT33s775KYpJOkUBJu0dxkJuUeIfDIOJRzAH084EQS3JN%2BzFjErLHuk%2FGZa9L4gBTSGCzeA69tI9PwLp7B37IQ%3D%3D&pageNo=1&numOfRows=10&LAWD_CD=11110&DEAL_YMD=201512")
#response는 url?ServiceKey&pageNo=1&numOfRows=10(item의 규격)&LAWD_CD&DEAL_YMD 를 합한 값을 가져오는 변수
response.encoding='utf-8'     # response의 인코딩 형식 지정
pageSource = response.text
xmlobj = BeautifulSoup(pageSource,"lxml-xml") #태그와 내용을 모두 보여줌

rowList = []                    # 행 내용을 담을 빈 배열을 생성
columnList = []                 # 열 내용을 담을 빈 배열을 생성
rows = xmlobj.findAll('item')   # "items" 안에 모든 내용을 담을 rows라는 변수를 생성

for i in range(0, len(rows)): 
    columns = rows[i].findAll() # 하나의 "item"의 모든 내용을 담을 columns라는 변수를 생성
    
    for j in range(0, len(columns)):  
        columnValue = columns[j].text    # item의 태그를 제외한 내용만을 담는 columnValue생성
        columnList.append(columnValue)   # 내용들을 columnList에 추가함
    rowList.append(columnList)           # 하나의 "item"이 columnList에 추가가 완료 되면 하나의 행(rows)이 완성됨 
    columnList = []                      # 다음 "item"값들을 넣어주기 위해 columnList를 초기화
    
homePriceSeoul = pd.DataFrame(rowList)  # 만들어진 표를 homePriceSeoul 변수에 담음
homePriceSeoul                          # 출력

columnHead = []               #item의 태그만을 받아 제목으로 쓰기위해 빈 리스트 columnHead를 만들어줌
for i in range(0, len(columns)):
    eachColumns = columns[i]        # item 하나당 받은 태그의 내용을 eachColumns 변수에 넣어줌
    columnHead.append(eachColumns.name)  # eachColumns의 태그이름만을 columnHead에 추가해줌
homePriceSeoul.columns = columnHead # homePriceSeoul테이블의 columns
homePriceSeoul                      # 출력
```
