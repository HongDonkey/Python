# 셀레니움을 활용하여 sparkkorea.com quiz에 접근을 한다.
# 이후 spark관련 퀴즈 타이틀과 링크정보를 수집하여 Df로 변환

```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup    # 필요한 라이브러리를 불러옴

baseUrl = "https://sparkkorea.com"   # 접속할 url을 변수에 담음
options = webdriver.ChromeOptions()  #  크롬에서 진행
driverLoc = "../addon/chromedriver/chromedriver.exe"   # 크롬드라이버 파일이 있는 경로를 변수로 지정
driver = webdriver.Chrome(executable_path=driverLoc,options=options) 
```

```python
driver.get(baseUrl)    # 페이지에 자동으로 접속
quizElementId = "menu-item-382"  # copy - selector 가져옴
driver.find_element_by_id(quizElementId).click() 
pgSource = driver.page_source # 페이지 구성을 불러옴
pgObj = BeautifulSoup(pgSource, "html.parser") # 개발자 페이지 코드를 bs4를 이용하여 출력
quiz=pgObj.find(name = "div", attrs={"id":"id_spark_quiz"}) # div클래스의 id_spark_quiz id를 가진 요소 출력
quiz2= quiz.find_all(name = "a") # a클래스 요소들을 변수에 담음
```

```python
titleList = []   # 제목을 담을 titleList 생성 
linkList = []    # 링크를 담을 linkList 생성
for i in range(0, len(quiz2)):
    title = quiz2[i].text     # A태그 안의 내용을 Title변수에 담음
    titleList.append(title)       # title변수를 titleList에 추가함
    link = quiz2[i].attrs["href"]  # A태그 안의 링크를 link변수에 담음
    linkList.append(link)       
    
```

```python
answer = pd.DataFrame(zip (titleList, linkList), columns=["퀴즈타이틀","링크"]) # 데이터프레임으로 출력
```

```python
answer.to_csv("./quiz_result.csv", index=False, encoding="ms949") # csv 파일 형태로 저장
```
